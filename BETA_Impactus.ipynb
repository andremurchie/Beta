```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web

data_inicio = ''
data_final = ''

benchmark = '^BVSP'
ticker = ''

#puxando cotações do ativo e do benckmark
cotacoes_ativo = web.DataReader('{}'.format(ticker), data_source='yahoo', start='{}'.format(data_inicio), end='{}'.format(data_final)) 
cotacoes_ibov = web.DataReader('{}'.format(benchmark), data_source='yahoo', start='{}'.format(data_inicio), end='{}'.format(data_final))

retorno_ativo = pd.DataFrame()
retorno_ibov = pd.DataFrame()
#calculo dos retornos do ativo e do benchmark
retorno_ativo['Retornos {}'.format(ticker)] = cotacoes_ativo['Adj Close'].pct_change()
retorno_ibov['Retornos Ibov'] = cotacoes_ibov['Adj Close'].pct_change()

ret_ativo_ibov = pd.DataFrame()
#unindo ('Merge') os dois dataframes
ret_ativo_ibov = retorno_ativo.merge(retorno_ibov, on='Date')
ret_ativo_ibov = ret_ativo_ibov.drop(ret_ativo_ibov.index[0])

#display(ret_ativo_ibov)

cov_ativo_ibov = pd.DataFrame()
#calculando covariância Ativo, IBOV
cov_ativo_ibov = np.cov(ret_ativo_ibov['Retornos {}'.format(ticker)], ret_ativo_ibov['Retornos Ibov'])[1][0]
print("Covariancia Ibov-Ativo: {}".format(cov_ativo_ibov))

#calculando variancia do IBOV
variancia_ibov = np.var(retorno_ibov['Retornos Ibov'])
print("Variancia_Ibov: {}".format(variancia_ibov))

#calculando beta
beta = cov_ativo_ibov / variancia_ibov
print("Beta: {}".format(beta))
```
